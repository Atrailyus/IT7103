# Slicing DataFrame

There are a few ways of slicing a Pandas DataFrame.

Using a **single index system** results in pandas trying to select **columns**. In dataframes that have headers, using numeric index results in error. You have to use the columns' names instead.

```python
data[0]
```

Slicing correctly, the selected column will be **viewed** as a 1D series

```python
data['YearsAtCompany']
```

We can select multiple columns in a list which result in a dataframe view

```python
data[['PerformanceRating','YearsAtCompany','YearsInCurrentRole']]
```

Row slicing can be added after columns in a <b>different</b> bracket system

```python
data[['PerformanceRating','YearsAtCompany','YearsInCurrentRole']][10:20]
```

To have a slicing syntax that is closer to numpy, we use `DataFrame.loc` or `DataFrame.iloc` to slice DataFrame. You need to provide two slicing conditions, one for row slicing and one for column slicing:

**`DataFrame.loc[{row condition} , {column condition}]`**    <== this is to slice columns using their names

or

**`DataFrame.iloc[{row condition} , {column condition}]`**   <== this is to slice columns using their numeric index

**<i>An important note when slicing a pandas dataframe with `loc` is that the `end_index` will now be included in the result</i>**

```python
data.loc[[10,60,110,200],['Department','EnvironmentSatisfaction','MonthlyIncome']]
```

```python
data.loc[:30:3,['Department','EnvironmentSatisfaction','MonthlyIncome']]
```

```python
data.loc[60:100:4,'YearsAtCompany':'YearsInCurrentRole']
```

`iloc` will give a slicing scheme that is identical to `numpy`

```python
#to use column indices instead of their names, we use iloc
data.iloc[[10,40,70,210,1004], [1,3,5,7]]
```

Being identical to numpy, the end_index will **not** be included. This is very important to remember, since incorrect index may lead to the target being included in the feature set which make any models meaningless.

To be certain, **always print** the slicing result so you can verify it is as you expect.

```python
data.iloc[:10, 4:-1]
```

Similar to a numpy array, we can slice a dataframe using conditions with both with loc and iloc. You have to make sure the columns are in their correct data types to avoid unexpected results or errors. For examples, compare an object column to a number may cause weird behaviours.

```python
data.loc[data['YearsAtCompany']<=30,:]
```

```python
data.loc[data['EnvironmentSatisfaction']=='Very High',:]
```

we can combine conditions, `&` for **and**, and `|` for **or**. Remember to wrap each condition in parentheses **()**

```python
data.loc[(data['YearsAtCompany']==10) & (data['MonthlyIncome']<=5100),:]
```

Note that slicing with `loc()` or `iloc()` give a view of the DataFrame.

We can make a copy if the particular subset is important to use later on.

```python
data_young_urban = data.loc[(data['EnvironmentSatisfaction']=='High') & (data['YearsInCurrentRole']<=7), :]
data_young_urban
```

## Binning Numeric Columns

Occasionally, the original values of a numeric column could be noisy which hurts a model's performance. That can sometimes be solved by **binning the column**. Examples are ages into age groups, incomes into income groups, and so on.

In this data, we can try binning `age`. First, I will reload the data to reset all changes.

```python
data = pd.read_csv('/content/drive/MyDrive/IT7103/employee_rating.csv')
data.head()
```

```python
data = pd.read_csv('/content/drive/MyDrive/IT7103/employee_general.csv')
data.head()
```

Then, we can use the `pandas.cut()` function to bin the column. This function takes two inputs:
- a list of split values. Each bin is formed from two consecutive splits in this list.
- a list of bin names. Should have one item less than the split list.

```python
bins = [0, 5, 60, np.inf] #the range are 0<age<=30, 30<age<=60, and 60<age
names = [1, 2, 3]          #age 0-30 will be bin 1, 30-60 bin 2, and over 60 bin 3

data['LessYears'] = pd.cut(data['YearsAtCompany'], bins=bins, labels=names)
data
```

Note that, categorizing a column should base on some justifications. For example, age groups into youth, middle age, elderly, or bmi into underweight, normal, and overweight. Arbitrary categories may be more harmful than helpful for a model.

## Convert Ordinal Categories to Numbers

It is also quite often we want to convert categories into numbers. However, you have to make sure the order in the numerical values makes sense.

For example, categories like `low`, `medium`, `high` can be converted to 1, 2, 3 because 1 < 2 is equivalent `low` < `medium` which is meaningful.

However, categories like `Europe`, `Asia`, `Africa`... do not have any inherent order. If you convert them to 1, 2, 3, does 1 < 2, or `Europe` < `Asia` make sense? These types of data need a different treatment discussed in module 5.

In this example, we will convert three groups of `smoking_status` into 1, 2, 3 as follows

- never smoked ==> 1
- formerly smoked ==> 2
- smokes ==> 3

Note that this conversion does not necessarily make senses, we are just using this column as an example. We will use `pandas.map()`.

`pandas.map()` takes input as a **dictionary** where each key-value pair representating the transformation of one category to one new value.

```python
Environment_map = {
    'High':1,
    'Medium':2,
    'Low':3
}
data['EnvironmentLevel'] = data['EnvironmentSatisfaction'].map(Environment_map)
data
```

There are multiple types of merging. If all  instances in either dataframe have matched IDs on the other, different types of merging give the same result. Otherwise, we need to understand our needs before merging to select the correct type.

For example, we want to merge two small dataframe as follows, but some rows do not have IDs in both sides

```python
demo_s = pd.read_csv('/content/drive/MyDrive/IT7103/demography_small.csv')
demo_s
```

```python
med_s = pd.read_csv('/content/drive/MyDrive/IT7103/medical_history_small.csv')
med_s
```

When calling `merge()`, the dataframe from which `merge()` is called is the **left** dataframe, and the one used as argument is the **right** dataframe.

For example,

`demo.merge(medhis, on='subject_id')`

means `demo` is the left dataframe, and `medhis` is the right dataframe. On the other hand,

`medhis.merge(demo, on='subject_id')`

means `medhis` is the left dataframe, and `demo` is the right one.

A **left-merge** keep all the rows in the left dataframe, and only rows in the right dataframe that have matched ID from the left. Rows from the left without matched IDs in the right have all their right columns become missing (NaN)

```python
demo_s.merge(med_s, on='subject_id', how='left')
```

A **right-merge** keep all the rows in the right dataframe, and only rows in the left dataframe that have matched ID from the right. Rows from the right without matched IDs in the right have all their left columns become missing (NaN)

```python
demo_s.merge(med_s, on='subject_id', how='right')
```

An **outer-merge** keeps all rows. Rows without matched from either sides will have those sections as missing

```python
demo_s.merge(med_s, on='subject_id', how='outer')
```

An **inner-merge** only keeps rows with matched IDs from both sides. All rows without matches are dropped.

```python
demo_s.merge(med_s, on='subject_id', how='inner')
```
